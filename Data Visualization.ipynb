```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib import rcParams
```

This is needed to ensure that axis labels are not cut off when exporting the graphs.

```python
rcParams.update({"figure.autolayout": True})
```

```python
data = pd.read_csv("clean_data.csv")
```

The majority of the people represented in the data set are middle-aged, with a large spike in the number of people in the 55-60 age range.

```python
sns.distplot(data.age)
```

There is not a significant correlation between education level and credit rating.

```python
sns.boxplot(data=data, x="education", y="credit_rating")
```

There is no relationship between marital status and household income.

```python
sns.boxplot(data=data, x="marital_status", y="household_income")
```

Building with a higher construction quality were typicall built more recently.

```python
sns.stripplot(
    data=data, x="construction_quality", y="year_built", jitter=True,
    order=["POOR", "FAIR", "AVERAGE", "GOOD", "EXCELLENT", "LUXURY"])
```

The average living area per household is the same in households with children as in households without children.

```python
sns.boxplot(data=data, x="children_in_household", y="living_area")
```

```python
sns.stripplot(data=data, x="bld_val", y="construction_type", jitter=True)
```

```python
sns.stripplot(data=data, x="bld_val", y="architecture_type", jitter=True)
```

There is no correlation bewteeen age and living area.

```python
plt.scatter(data.age, data.living_area)
```

```python
sns.distplot(data.living_area)
```

```python
plot = sns.pointplot(data=data, x="age", y="net_worth", hue="children_in_household", ci=False, scale=0.5)
plot.xaxis.set_major_locator(ticker.MultipleLocator(10))
plot.xaxis.set_major_formatter(ticker.ScalarFormatter())
```

```python
plot = sns.lmplot(data=data, x="age", y="net_worth", hue="children_in_household", scatter=False)
```
