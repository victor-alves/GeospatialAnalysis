```python
import pandas as pd
import numpy as np
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
rcParams.update({"figure.autolayout": True})
```

```python
sns.set(style="white", color_codes=True)
```

```python
df = pd.read_csv("clean_data.csv")
df.head(1)
```

```python
f, (ax1, ax2, ax3, ax4) = plt.subplots(4)
```

```python
archs = df["architecture_type"].value_counts().values.tolist()
names = df["architecture_type"].value_counts().index.tolist()
indexes = np.asarray(list(range(len(archs))))

df2 = pd.DataFrame({"architecture_type": archs, "names": names})
df2.set_index(indexes, inplace=True)

z = sns.barplot(y="architecture_type", x="names", ci="sd", errwidth=.8, data=df2)
z.set_xticklabels(labels=names, rotation=90)
```
